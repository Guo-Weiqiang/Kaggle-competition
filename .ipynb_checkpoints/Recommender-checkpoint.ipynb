```python
def transfom_date_time(data):
    data['date_time'] = pd.to_datetime(data['date_time'])
    data['season'] = data['date_time'].dt.quarter
    
    data.drop(labels=["date_time"], axis=1, inplace=True)
    
    return data
```

```python
def remove_columns(data, ignore_column=["srch_id", "prop_id", "position", "random_bool"]):
    ignore_column = [c for c in ignore_column if c in data.columns.values]

    data = data.drop(labels=ignore_column, axis=1)

    return data

def compute_estimated_position(training_data, srch_id_dest_id_dict):
    training_data = training_data.merge(
    srch_id_dest_id_dict, how="left", on=["srch_destination_id", "prop_id"] )

    return training_data

def train_val_last_process(x1, y1, x2, y2):
    # compute estimated position for training set and validation set
#     srch_id_dest_id_dict = x1.loc[x1["random_bool"] == 0]
    srch_id_dest_id_dict = x1
    srch_id_dest_id_dict = x1.groupby(["srch_destination_id", "prop_id"]).agg({"position": "mean"})
    srch_id_dest_id_dict = srch_id_dest_id_dict.rename(index=str, columns={"position": "estimated_position"}).reset_index()
    srch_id_dest_id_dict["srch_destination_id"] = (srch_id_dest_id_dict["srch_destination_id"].astype(str).astype(int))
    srch_id_dest_id_dict["prop_id"] = (srch_id_dest_id_dict["prop_id"].astype(str).astype(int))
    srch_id_dest_id_dict["estimated_position"] = (1 / srch_id_dest_id_dict["estimated_position"])
    
    x1 = compute_estimated_position(x1, srch_id_dest_id_dict)
    x2 = compute_estimated_position(x2, srch_id_dest_id_dict)
    
    groups = x1["srch_id"].value_counts(sort=False).sort_index()
    eval_groups = x2["srch_id"].value_counts(sort=False).sort_index()

    x1 = remove_columns(x1)
    x2 = remove_columns(x2)
    
    
    return (x1, x2, y1, y2, groups, eval_groups, srch_id_dest_id_dict)
```

```python
def add_mean_features(data, attributes, group):
    for attr in attributes:
        mean_per_id = data.groupby(group)[attr].mean()

        index = data[group]
        new_attr = mean_per_id[index].tolist()

        name = attr + '_avg_' + group
#         data[name] = new_attr
        data[name] = data[group] - new_attr # attr substracts its mean value
        
        
    return data
```

```python
def add_median_features(data, attributes, group):
    for attr in attributes:
        mean_per_id = data.groupby(group)[attr].median()

        index = data[group]
        new_attr = mean_per_id[index].tolist()

        name = attr + '_median_' + group
#         data[name] = new_attr
        data[name] = data[group] - new_attr
        
    return data
```

```python
def add_std_features(data, attributes, group):
    for attr in attributes:
        std_per_id = data.groupby(group)[attr].std()

        index = data[group]
        new_attr = std_per_id[index].tolist()

        name = attr + '_std_' + group
        data[name] = new_attr
        
    return data
```

```python
def add_listwise_rank(data, rank_columns):
    for attr in rank_columns:
        name = attr + '_rank'
        data[name] = data.groupby("srch_id")[attr].rank(method="first", ascending=True)
    
    return data
```

```python
def add_composite_features(data):
    data['hist_price_diff'] = data['prop_log_historical_price'] - data['price_usd']
    data['price_diff'] = data['visitor_hist_adr_usd'] - data['price_usd']
    data['prop_starrating_diff'] = data['visitor_hist_starrating'] - data['prop_starrating']
        
    return data
```

```python
def normalize_features(data, group_key, target_column, take_log10=False):
    epsilon = 1e-4
    
    if take_log10:
        data[target_column] = np.log10(data[target_column] + epsilon)
        
    methods = ["mean", "std"]
    df = data.groupby(group_key).agg({target_column: methods})
    df.columns = df.columns.droplevel()
    col = {}
    
    for method in methods:
        col[method] = target_column + "_" + method
        
    df.rename(columns=col, inplace=True)
    df_merge = data.merge(df.reset_index(), on=group_key)
    df_merge[target_column + "_norm_by_" + group_key] = (df_merge[target_column] - df_merge[target_column + "_mean"]) / df_merge[target_column + "_std"]
    df_merge = df_merge.drop(labels=[col["mean"], col["std"]], axis=1)

    return df_merge
```

```python
def add_normalized_features(data):
#     data = normalize_features(data.copy(),group_key="srch_id",target_column="price_usd",take_log10=True)
#     data = normalize_features(data.copy(),group_key="prop_id",target_column="price_usd")
#     data = normalize_features(data.copy(),group_key="srch_destination_id",target_column="price_usd")
    
#     prop_num_attr = ['prop_review_score', 'prop_location_score2', 'srch_query_affinity_score']
#     groups = ['srch_id', 'prop_id', 'srch_destination_id']
    
#     for attr in prop_num_attr:
#         for group in groups:
#             data = normalize_features(data, group_key=group, target_column=attr)          
    data = normalize_features(data, group_key="srch_id", target_column="price_usd", take_log10=True)
    
    data = normalize_features(data, group_key="prop_id",target_column="price_usd")
    data = normalize_features(data, group_key="srch_destination_id",target_column="price_usd") 
       
    data = normalize_features(data, group_key="srch_id", target_column="prop_starrating")
#     data = normalize_features(data, group_key="prop_id", target_column="prop_starrating")
    data = normalize_features(data, group_key="srch_destination_id", target_column="prop_starrating")
    data = normalize_features(data, group_key="srch_id", target_column="prop_location_score2")
    data = normalize_features(data, group_key="prop_id", target_column="prop_location_score2") #ï¼Ÿ
    data = normalize_features(data, group_key="srch_destination_id", target_column="prop_location_score2")
    data = normalize_features(data, group_key="srch_id", target_column="prop_location_score1")
#     data = normalize_features(data, group_key="prop_id", target_column="prop_location_score1")
    data = normalize_features(data, group_key="srch_destination_id", target_column="prop_location_score1")
    data = normalize_features(data, group_key="srch_destination_id", target_column="prop_review_score")
#     data = normalize_features(data, group_key="prop_id", target_column="prop_review_score")
    data = normalize_features(data, group_key="srch_id", target_column="prop_review_score")
    

    data = data.sort_values("srch_id")
    
    return data
```

## Feature Engineering

```python
def compute_relavance(data):
    relavance_base = data['click_bool'] + data['booking_bool']
    
    relavance = []
    for i in relavance_base:        
        if i == 2:
            relavance.append(5)
        elif i == 1:
            relavance.append(1)
        else:
            relavance.append(0)

    data['relavance'] = relavance
        
    return data
```

```python
# remove outliers
def remove_outliers(data, columns):
    for col in columns:
        q75 = data[col].quantile(0.75)
        q25 = data[col].quantile(0.25)
        intr_qr = q75-q25
        
        max = q75+(1.5*intr_qr)
        min = q25-(1.5*intr_qr)
        
        index = data[data[col] > max].index.tolist() 
        #+ data[data[col] < min].index.tolist()
        data.drop(index, inplace=True)
        
    return data
```

```python
def drop_missing_data(data, threshold):
    ignore = ['visitor_hist_starrating', 'visitor_hist_adr_usd', 'srch_query_affinity_score']
    
    null_sum = data.isnull().sum()
    drop_columns = data.columns[null_sum > len(data) * threshold].tolist()
    drop_columns = [drop_columns[i] for i in range(len(drop_columns)) if drop_columns[i] not in ignore]
    print('Columns {} have been dropped'.format(drop_columns))
    
    data.drop(columns=drop_columns, inplace=True)
    
    return data
```

```python
def data_preprocessing(data, drop_columns, outlier_columns, type="train"):
    if type != "test":
        data = compute_relavance(data)
    
    # remove features
    data.drop(columns=drop_columns, inplace=True)
    
    # remove missing features
    data = drop_missing_data(data, 0.4)
    
    # filling the missing features
    data['prop_review_score'].fillna(-1, inplace=True)
    data['prop_location_score2'].fillna(-1, inplace=True)
#     data['srch_query_affinity_score'].fillna(data['srch_query_affinity_score'].median(), inplace=True)
    data['srch_query_affinity_score'].fillna(-1, inplace=True)
#     data['orig_destination_distance'].fillna(data['orig_destination_distance'].median(), inplace=True)
    data['orig_destination_distance'].fillna(-1, inplace=True)
    data['visitor_hist_starrating'].fillna(-1, inplace=True)
    data['visitor_hist_adr_usd'].fillna(-1, inplace=True)
#     data['visitor_hist_starrating'].fillna(data['visitor_hist_starrating'].median(), inplace=True)
#     data['visitor_hist_adr_usd'].fillna(data['visitor_hist_adr_usd'].median(), inplace=True)
    
    # add time features
    data = transfom_date_time(data)

#     # add mean, median, std features
    agg_attrs = ['prop_starrating', 'prop_review_score', 'prop_location_score1', 'prop_location_score2']
    data = add_mean_features(data, agg_attrs, 'prop_id')
    data = add_median_features(data, agg_attrs, 'prop_id')
    data = add_std_features(data, agg_attrs, 'prop_id')
#     data = add_mean_features(data, agg_attrs, 'srch_id')
#     data = add_median_features(data, agg_attrs, 'srch_id')
#     data = add_std_features(data, agg_attrs, 'srch_id')
#     data = data.sort_values("srch_id")
    
    # add normalized_features
    data = add_normalized_features(data)
    
    # add composite_features
    data = add_composite_features(data)
    
    # add listwise(rank) features
    rank_columns = ['price_usd', 'prop_starrating', 'prop_location_score2', 'price_diff']
    data = add_listwise_rank(data, rank_columns)

    # remove 
    drop_columns = ["price_usd"]
# #     drop_columns = ["price_usd", "prop_starrating", "prop_location_score2", "prop_location_score1", "prop_review_score"]
    data.drop(columns=drop_columns, inplace=True)
    
    return data
```

```python
def predict(test_set, srch_id_dest_id_dict):
    test_set = compute_estimated_position(test_set, srch_id_dest_id_dict)
    test_set_srch_id_prop_id = test_set[["srch_id", "prop_id"]]
    test_set = remove_columns(test_set)
    

    print("Predicting on train set with columns: {}".format(test_set.columns.values))

    
    predictions = ranker.predict(test_set)
    test_set_srch_id_prop_id["prediction"] = predictions

    test_set_srch_id_prop_id = test_set_srch_id_prop_id.sort_values(
            ["srch_id", "prediction"], ascending=(True, False)
        )
    
    print("Saving predictions into new_submission.csv")
    test_set_srch_id_prop_id[["srch_id", "prop_id"]].to_csv(path_or_buf='new_submission', index=False)
    
    return test_set
```
